```python
#71
numero = int(input("Digite o número: "))

if(numero < 6) and (numero > -6):
  print("Não é triangular")
else:
  multiplicacao = 1
  triangularValidade = False
  i = 1
  while i <= numero and triangularValidade == False:
    if numero > 0:
     for j in range(3):
        multiplicacao = multiplicacao * (i+j)

        if(multiplicacao == numero):
            triangularValidade = True
        else:
            i += 1
            multiplicacao = 1
    if numero < 0:
        i = -1
        for j in range(3):
            multiplicacao = multiplicacao * (i+j)

            if(multiplicacao == numero):
                triangularValidade = True
            else:
                i -= 1
                multiplicacao = 1

  if(triangularValidade == True):
    print("O número é triangulara")
  else:
    print("O número não é triangular")
```
